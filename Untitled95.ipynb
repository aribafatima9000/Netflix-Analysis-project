```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df.head()
```

# DATA CLEANING

# DROP COLUMNS

```python
df.drop(columns = ["listed_in","description","show_id"],inplace = True)
```

```python
df.head()
```

# CHECK DATATYPES

```python
df.info()
```

```python
df['release_year'] = pd.to_datetime(df['release_year'], errors='coerce')
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.info()
```

# CHECK NULL VALUES

```python
df.isnull().sum()
```

```python
df["director"].fillna("unknow", inplace=True)
df["cast"].fillna("unknow", inplace=True)
df["country"].fillna("unknow", inplace=True)
```

```python
df["country"].fillna(method="ffill", inplace=True)
df["rating"].fillna(method="ffill", inplace=True)
df["duration"].fillna(method="ffill", inplace=True)
df['release_year'] = df['release_year'].dt.year  # sirf year extract karo
df['release_year'].fillna(2021, inplace=True)    # missing values ko default 2021 se replace
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# PROBLEMS STATMENTS 

#Content Type Distribution

```python
import matplotlib.pyplot as plt
type_counts = df['type'].value_counts()
type_percent = df['type'].value_counts(normalize=True) * 100

print("📊 Content Type Count:\n", type_counts)
print("\n📊 Content Type Percentage:\n", type_percent)

## ---------------------------------------
# Visualization using matplotlib only
# ---------------------------------------
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values, color=['red', 'yellow'])
plt.title('Movies vs TV Shows Distribution')
plt.xlabel('Content Type')
plt.ylabel('Count')

# Count labels upar show karna
for i, v in enumerate(type_counts.values):
    plt.text(i, v + 0.1, str(v), ha='center', fontsize=12)

plt.show()
```

```python
#Movies ka count ziada (6131) aur percentage 69%+ hai, iska matlab platform par ziyada movies release hui hain.

#TV Shows ka percentage kam (30%) hai, matlab movies platform par dominate kar rahi hain.
```

#Country-wise Content Analysis

```python
import matplotlib.pyplot as plt

# Country count
country_counts = df['country'].value_counts()
print(" Country-wise Content Count:\n", country_counts)

```

```python
import matplotlib.pyplot as plt

# Unknown hatao
df_director = df[df['director'] != 'unknow']

# Count
top_directors = df_director['director'].value_counts().head(10)

print("🎬 Top 10 Directors:\n", top_directors)

# Plot
plt.figure(figsize=(10,5))
plt.bar(top_directors.index, top_directors.values)

plt.title('Top 10 Directors by Content Count')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=60)

# Values show
for i, v in enumerate(top_directors.values):
    plt.text(i, v + 0.2, str(v), ha='center')

plt.show()
```

```python
# sub sa zayda rajiv chilaka ha as ki movies sub a zayda acha ha 
```

```python
# Cast column ko split karna (kyunki multiple actors hain)
df_cast = df[df['cast'] != 'unknow'].copy()

# Split actors
df_cast['cast'] = df_cast['cast'].str.split(',')

# Explode (har actor alag row me)
df_cast = df_cast.explode('cast')

# Spaces remove
df_cast['cast'] = df_cast['cast'].str.strip()

# Count top actors
top_actors = df_cast['cast'].value_counts().head(10)

print("⭐ Top 10 Actors:\n", top_actors)

# Plot
plt.figure(figsize=(10,5))
plt.bar(top_actors.index, top_actors.values)

plt.title('Top 10 Actors by Appearances')
plt.xlabel('Actors')
plt.ylabel('Number of Appearances')
plt.xticks(rotation=60)

# Values show
for i, v in enumerate(top_actors.values):
    plt.text(i, v + 0.2, str(v), ha='center')

plt.show()
```

```python
# top actor ya ha anupam kher as movies ma rol bohot acha haa 
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Sirf Movies filter karo
df_movies = df[df['type'] == 'Movie']

# 2️⃣ Unknown remove karo
df_movies = df_movies[df_movies['director'] != 'unknow']

# 3️⃣ Sirf top 10 movies lo (graph clean rahe)
df_sample = df_movies.head(10)

# 4️⃣ Graph
plt.figure(figsize=(10,6))

plt.barh(df_sample['title'], range(len(df_sample)))

plt.title('Movies with Their Directors')
plt.xlabel('Index')
plt.ylabel('Movie Name')

# Director name show karna
for i, director in enumerate(df_sample['director']):
    plt.text(0, i, director, fontsize=9)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 50 longest movies dikhayenge (readable)
top_movies = df_movies.sort_values(by='duration', ascending=False).head(50)

plt.figure(figsize=(12,10))
plt.barh(top_movies['title'], top_movies['duration'], color='skyblue')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Movie Name')
plt.title('Top 50 Longest Movies')

# Values show karna
for i, v in enumerate(top_movies['duration']):
    plt.text(v + 1, i, str(v), fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Sirf Movies filter
df_movies = df[df['type'] == 'Movie'].copy()

# 2️⃣ Sirf 'min' wali durations
df_movies = df_movies[df_movies['duration'].str.contains('min')]

# 3️⃣ Duration ko numeric me convert karo
df_movies['duration'] = df_movies['duration'].str.replace(' min', '').astype(int)

# 4️⃣ Top 50 shortest movies
shortest_movies = df_movies.sort_values(by='duration', ascending=True).head(50)

print("🎬 Top 50 Shortest Movies:\n", shortest_movies[['title','duration']])

# 5️⃣ Horizontal Bar Graph
plt.figure(figsize=(12,10))
plt.barh(shortest_movies['title'], shortest_movies['duration'], color='salmon')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Movie Name')
plt.title('Top 50 Shortest Movies')

# Values show karna
for i, v in enumerate(shortest_movies['duration']):
    plt.text(v + 0.5, i, str(v), fontsize=8)

plt.tight_layout()
plt.show()
```
